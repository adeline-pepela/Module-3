```python
import networkx as nx
import matplotlib.pyplot as plt
import community # Python-Louvain
import numpy as np
```

```python
# Create undirected graph
G = nx.Graph()
```

```python
# Create nodes
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
```

```python
# Draw graph
nx.draw(G, with_labels=True, font_weight='bold', node_color='skyblue', node_size=1500, font_size=8)
plt.show()
```

```python
# Create directed graph
G = nx.DiGraph()
```

```python
# Create nodes
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
```

```python
# Create adjacency matrix
adj_matrix = nx.adjacency_matrix(G).todense()
print(adj_matrix)
```

```python
# Draw graph
nx.draw(G, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
plt.show()
```

```python
# Using node fucntiom
asj_list = {node:list(G.neighbors(node)) for node in G.nodes}
print(asj_list)
```

```python
G = nx.Graph()
edges = [(1, 2, 7), (1, 9, 3), (1, 6, 14), (2, 3, 10),(2, 4, 15),(3, 4, 11),(3, 6, 2),(4, 5, 6),(5, 6, 9)]
G.add_weighted_edges_from(edges)
```

```python
# Shortest path
shortest_path = nx.shortest_path(G, source=1, target=5, weight='weight')
print(shortest_path)
```

```python
# Draw graph
nx.draw(G, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
plt.show()
```

```python
# Create graph using karate
G = nx.karate_club_graph()

# Community partition using Louvain
partition = community.best_partition(G)
print(partition)
```

```python
# Draw graph
plt.figure(figsize=(10, 10))
```

```python
nx.draw(G, pos=nx.spring_layout(G), cmap=plt.get_cmap('jet'), node_color=list(partition.values()), node_size=300, with_labels=True, font_size=10, font_color='white')
plt.show()
```
