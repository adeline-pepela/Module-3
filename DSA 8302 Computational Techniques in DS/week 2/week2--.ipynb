```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###Random walk
num_step=1000 #Number of steps
steps=np.random.choice([-1,1], size=num_step) #Random choice -1 and 1
position=np.cumsum(steps) #Cumulative sum to get position
```

```python
#plot
plt.figure(figsize=(10,5))
plt.plot(position, label="1d Random Walk")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('1D Random Walk')
plt.legend()
plt.show()
```

```python
#Compare multiple 1D Random Walks (Comparing 5 walks)
plt.figure(figsize=(10,6))
for i in range(5):
    steps=np.random.choice([-1,1], size=num_step) #Random Choice -1 and 1
    position=np.cumsum(steps) #Cumulative sum to get position
    plt.plot(position, alpha=0.75)
```

```python
num_steps=1000
x_steps=np.random.choice([-1,1], size=num_steps) #steps in x-direction
y_steps=np.random.choice([-1,1], size=num_steps) #steps in y-direction
```

```python
x_positions=np.cumsum(x_steps) #Cumulative sum to get position in x-direction
y_positions=np.cumsum(y_steps) #Cumulative sum to get position in y-direction
```

```python
###2D random Walk
plt.figure(figsize=(10,6))
plt.plot(x_positions, y_positions, marker='.', linestyle="-", alpha=0.7)
plt.scatter(0, 0, color='red', label="Start", zorder=3)
plt.scatter(x_positions[-1], y_positions[-1], color='blue', label="End", zorder=3)
plt.xlabel('x position')
plt.ylabel('y position')
plt.title('2D Random Walk')
plt.legend()
plt.grid()
plt.show()
```

```python
#Higher probability for positive direction
prob_right=0.7 #70% chance of moving to the right direction
steps=np.random.choice([-1,1], size=num_steps, p=[1-prob_right, prob_right])
position=np.cumsum(steps)
```

```python
plt.figure(figsize=(10,5))
plt.plot(position, label="Adjusted 1D Random Walk")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Adjusted 1D Random Walk')
plt.legend()
plt.show()
```
