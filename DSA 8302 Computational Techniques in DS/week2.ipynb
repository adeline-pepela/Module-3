```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#random with 10
num_step=1000 #number of steps
steps=np.random.choice([-1,1], size=num_step) #random choice -1 and 1
position=np.cumsum(steps) #cumulative sum to get position
```

```python
#plot
plt.figure(figsize=(10,5))
plt.plot(position,label="ID Random Walk")
plt.axhline(y=0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel("Steps")
plt.ylabel("Position")
plt.title("ID Random Walk")
plt.legend()
plt.show()
```

```python
#compare multple ID Random Walk 
plt.figure(figsize=(10,6))
for i in range(5):
    steps=np.random.choice([-1,1], size=num_step) #Random choice -1 and 1
    position=np.cumsum(steps) #cumulative sum to get position
    plt.plot(position, alpha=0.75)
```

```python
num_steps=1000
x_steps=np.random.choice([-1,1],size=num_steps) #steps in x- direction
y_steps=np.random.choice([-1,1],size=num_steps) #steps in y- direction
```

```python
x_positions=np.cumsum(x_steps)
y_positions=np.cumsum(y_steps)
```

```python
#2D Random walk 
plt.figure(figsize=(7,7))
#plt.plot(x_positions, y_positions, marker=".", linestyle="__", alpha=0.7)
plt.scatter(0,0,color="red", label="Start", zorder=3)
plt.scatter(x_positions[-1], y_positions[-1], color="blue", label="End", zorder=3)
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("2D Random Walk")
plt.legend()
plt.grid()
plt.show()
```

```python
#higher probability for positive directions
prob_right=0,7 #70 % chance of moving to the right direction 
steps=np.random.choice([-1,1], size=num_steps, p=[1-prob_right,prob_right])
position=np.cumsum(steps)
```

```python
plt.figure(figsize=(10,5))
plt.plot(position, label="Adjusted ID Random Walk")
plt.axhline(y=0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel("Steps")
plt.ylabel("Position")
plt.title(" Adjusted ID Random Walk")
plt.legend()
plt.show()
```
